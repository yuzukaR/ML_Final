```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
url = '/Users/rachelchen/Desktop/ML/project/housing.csv'
df = pd.read_csv(url)
df.head()
```

```python
median_value = df['total_bedrooms'].median()
df['total_bedrooms'] = df['total_bedrooms'].fillna(median_value)
```

```python
df.isnull().sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is the DataFrame containing your dataset

# Define the price ranges, with a bin width of 500
bins = range(14999, 500000, 500)  # From 14999 to 500000, with a step of 500
labels = [f"${i}-{i+500}" for i in range(14999, 500000, 500)]  # Labeling each range

# Remove the last label to match the number of bins
labels = labels[:-1]

# Create a new column in the DataFrame for price ranges
df['price_range'] = pd.cut(df['median_house_value'], bins=bins, labels=labels, right=False)

# Plot the distribution of the price ranges
plt.figure(figsize=(12, 6))
df['price_range'].value_counts().sort_index().plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Price Range (in $500 intervals)')
plt.ylabel('Number of Houses')
plt.title('Distribution of House Prices by $500 Range')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
